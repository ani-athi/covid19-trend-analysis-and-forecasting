# Filtering and grouping

```python
global_daily = (
    covid_19
    .groupby('Date')[['Confirmed','Deaths','Recovered','Active']]
    .sum()
    .reset_index()
    .sort_values('Date')
)
global_daily
```

```python
print(global_daily.shape)
```

```python
country_daily = covid_19.groupby(['Date', 'Country/Region'])[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()
country_daily
```

# Visualization

```python
import plotly.express as px
```

```python
type(global_daily)
```

```python
fig = px.line(
    global_daily,
    x = 'Date',
    y = ['Confirmed', 'Recovered', 'Deaths'],
    title='Global COVID-19 Trend',
    labels={'value': 'Number of Cases', 'variable': 'Category'}
)

fig.update_layout(template = 'plotly_white')
fig.show()
```

```python



country_daily = (
    covid_19.groupby(['Date', 'Country/Region'])[['Confirmed', 'Deaths', 'Recovered', 'Active']]
    .sum()
    .reset_index()
    .sort_values('Date')
)

latest_totals = country_daily[country_daily['Date'] == country_daily['Date'].max()]
top_countries = latest_totals.sort_values('Confirmed', ascending=False)['Country/Region'].head(10).tolist()

subset = country_daily[country_daily['Country/Region'].isin(top_countries)]


fig = px.line(
    subset,
    x='Date',
    y='Confirmed',
    color='Country/Region',
    title = f'Confirmed Cases in Top{len(top_countries)} Countries',
    labels={'Confirmed': 'Confirmed cases'}
)

fig.update_layout(template='plotly_white', legend_title='Country')
fig.show()
```
